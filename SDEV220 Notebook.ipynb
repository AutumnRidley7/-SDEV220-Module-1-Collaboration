# Chapter 3 Assignment - Sections 3.1 to 3.6

## Section 3.1
*Prompt: How many seconds are in an hour?*
*Answer: There are 60 seconds in a minute and 60 minutes in an hour, so multiply 60 * 60.*

```python
seconds_in_hour = 60 * 60
seconds_in_hour
```

## Section 3.2
*Prompt: Assign the result from the previous task to a variable called seconds_per_hour.*
*Answer: Use the previous calculation and assign it to a variable.*

```python
seconds_per_hour = seconds_in_hour
seconds_per_hour
```

## Section 3.3
*Prompt: How many seconds are in a day? Use your seconds_per_hour variable.*
*Answer: Multiply seconds_per_hour by the number of hours in a day (24).*

```python
seconds_in_day = seconds_per_hour * 24
seconds_in_day
```

## Section 3.4
*Prompt: Calculate seconds per day again, but this time save the result in a variable called seconds_per_day.*
*Answer: Same as above, but store in seconds_per_day.*

```python
seconds_per_day = seconds_per_hour * 24
seconds_per_day
```

## Section 3.5
*Prompt: Divide seconds_per_day by seconds_per_hour. Use floating-point division.*
*Answer: Floating-point division uses / in Python.*

```python
seconds_per_day / seconds_per_hour
```

## Section 3.6
*Prompt: Divide seconds_per_day by seconds_per_hour, using integer division. Did this number agree with the floating-point value from the previous question, aside from the final .0?*
*Answer: Integer division uses // and it should give the same whole number result as floating-point division.*

```python
seconds_per_day // seconds_per_hour
```
